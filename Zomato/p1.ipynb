```python
# Now we have country code in both dataset
# so lets combine the two tables
final_df=pd.merge(df,df_country,on='Country Code',how='left') #how=left means left join like in sql
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
unique_values=final_df['Country'].unique()
print(unique_values) # gives name of each unique value in column


final_df.Country.value_counts()  # gives total count of each unique value in series
```

So from above we come to conclusion that in india zomato has max operability 

```python
y=final_df.Country.value_counts().index    # all country names in array
```

```python
x=final_df.Country.value_counts().values    # all value couunts of each counttry from value_counts
```

to see which country has maximum orders using pyplot

```python
# Pie Chart
plt.pie(x[:3],labels=y[:3],autopct="%2.2f%%")
plt.show()
```

observation: Zomato has maximum transactions or orders from India after that USA then UK
